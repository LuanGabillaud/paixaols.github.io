# Pandas 1

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

## Gr√°fico de linha

```python
np.random.seed(73)
x = np.arange(0, 10, 0.2)
y = np.random.randint(low=0, high=11, size=50)
df = pd.DataFrame({'col x':x, 'col y':y})
df.head()
```

```python
fig, ax = plt.subplots(figsize=(6,5))

df.plot(ax = ax)

fig.tight_layout()
# fig.savefig('../images/pandas1-1.png', facecolor='w', dpi=200)
```

```python
fig, ax = plt.subplots(figsize=(6,5))

df.plot(
    x='col x', y='col y',
    xlabel='Eixo x',
    ylabel='Eixo y',
    title='Gr√°fico de linha',
    ax=ax
)

fig.tight_layout()
# fig.savefig('../images/pandas1-2.png', facecolor='w', dpi=200)
```

## Histograma

```python
df.plot(
    y='col y',
    kind='hist',
    title='Histograma'
);
```

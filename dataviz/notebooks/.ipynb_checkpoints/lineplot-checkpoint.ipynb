# Lineplot 1
BÃ¡sico

```python
import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(6,5))

x = [1,2,3,4,5]
y = [2,5,3,5,7]

plt.plot(x, y)

fig.tight_layout()
# fig.savefig('../images/lineplot1-1.png', facecolor='w', dpi=200)
```

```python
fig = plt.figure(figsize=(6,5))

x = [1,2,3,4,5]
y = [2,5,3,5,7]

plt.plot(x, y, linestyle='--', marker='o')

fig.tight_layout()
# fig.savefig('../images/lineplot1-2.png', facecolor='w', dpi=200)
```

```python
fig = plt.figure(figsize=(6,5))

x = [1,2,3,4,5]
y = [2,5,3,5,7]

plt.plot(x, y, linestyle='--', marker='o')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.grid(linestyle='--')

fig.tight_layout()
# fig.savefig('../images/lineplot1-3.png', facecolor='w', dpi=200)
```

```python
fig = plt.figure(figsize=(6,5))

x = [1,2,3,4,5]
y1 = [2,5,3,5,7]
y2 = [7,4,5,3,2]

plt.plot(x, y1, linestyle='--', marker='o', label='Linha 1')
plt.plot(x, y2, linestyle='-.', marker='s', label='Linha 2')
plt.xlabel('Eixo x')
plt.ylabel('Eixo y')
plt.grid(linestyle='--')
plt.legend()

fig.tight_layout()
# fig.savefig('../images/lineplot1-4.png', facecolor='w', dpi=200)
```

# Lineplot 2

## ggplot

```python
import matplotlib.pyplot as plt
plt.style.use('ggplot')
```

```python
fig = plt.figure(figsize=(6,5))

x = [1,2,3,4,5]
y1 = [2,5,3,5,7]
y2 = [7,4,5,3,2]

plt.plot(x, y1, '-o', label='Linha 1')
plt.plot(x, y2, '-s', label='Linha 2')
plt.legend()
plt.title('ggplot')

fig.tight_layout()
# fig.savefig('../images/lineplot2-1.png', facecolor='w', dpi=200)
```
